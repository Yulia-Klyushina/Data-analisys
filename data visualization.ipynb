```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing distribution of data

Plot a histogram of `ConvertedComp.`

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
df_age = pd.read_sql_query(QUERY,conn)
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """
SELECT Age, WorkWeekHrs
FROM master

"""
df2 = pd.read_sql_query(QUERY,conn)
ax2 = sns.regplot(x = "Age", y = "WorkWeekHrs", data = df2)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master

"""
df3 = pd.read_sql_query(QUERY,conn)
df3
```

```python
norm_age = (df3['Age'] - df3['Age'].min()) / (df3['Age'].max() - df3['Age'].min())
ax0 = df3.plot(kind = 'scatter', x = 'WorkWeekHrs', y = 'CodeRevHrs', figsize =(14, 8), alpha = 0.5, color = 'green',
              s = norm_age * 1000, xlim = (0, 100))
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT DatabaseDesireNextYear, count(*) as count
FROM DatabaseDesireNextYear
Group by DatabaseDesireNextYear
Order by count desc limit 5
"""
df4 = pd.read_sql_query(QUERY,conn)
df4.set_index('DatabaseDesireNextYear',inplace=True)
df4
```

```python
df4['count'].plot(kind = 'pie',
                                  figsize = (5, 6),
                                  autopct = '%1.1f%%',
                                  startangle = 90)
plt.title('The top 5 databases that respondents wish to learn next year')
plt.show()
```

```python
QUERY = """
SELECT LanguageDesireNextYear, count(*) as count
FROM LanguageDesireNextYear
Group by LanguageDesireNextYear
Order by count desc limit 5
"""
df4 = pd.read_sql_query(QUERY,conn)
df4.set_index('LanguageDesireNextYear',inplace=True)
df4
```

```python
QUERY = """
SELECT count(*) as count
FROM LanguageWorkedWith
where LanguageWorkedWith = 'SQL'
"""
df4 = pd.read_sql_query(QUERY,conn)
df4
```

```python
QUERY = """
SELECT Respondent, DatabaseWorkedWith, count(DatabaseWorkedWith)
FROM DatabaseWorkedWith
group by Respondent
having count(DatabaseWorkedWith) = 1 and DatabaseWorkedWith = 'MySQL'
"""
df4 = pd.read_sql_query(QUERY,conn)
df4
df4['DatabaseWorkedWith'].value_counts()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age >= 30 and Age <= 35

"""
df5 = pd.read_sql_query(QUERY,conn)
df_median = df5.groupby('Age').median()
df_median
```

```python
df_median.plot(kind='bar',stacked=True)
plt.title('Median workweek hrs and CodeRevHrs in 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.show()


```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age >= 45 and Age <= 60

"""
df6 = pd.read_sql_query(QUERY,conn)
df_median = df6.groupby(['Age'], as_index = False).median()
df_median
```

```python
plt.plot(df_median['Age'], df_median['ConvertedComp'], color='red', marker='o')
plt.title('Median ConvertedComp for all ages from 45 to 60')
plt.xlabel('Age')
plt.ylabel('USD')
plt.show()
```

Create a horizontal bar chart using column `MainBranch.`
